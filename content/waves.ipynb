# Ondas

## Ondas a uma Dimensão

Vamos começar por descrever ondas que se propagam a uma dimensão, como por exemplo ondas transversais numa corda tensa ou então ondas longitudinais numa mola (ou até o som que se propaga num tubo cilíndrico).

Para descrever uma onda que se propaga ao longo de um meio unidimensional podemos usar a seguinte expressão:

$$
y(x, t) = A \sin\left[\frac{2\pi}{\lambda}(x-vt)\right]
$$
